## 0)— Day 81 Goals: Multivariable Regression – Predict House Prices

### Introduction

Welcome to Boston, Massachusetts in the 1970s.

You are working for a real estate development company that needs to estimate the value of residential projects before construction begins.

Your task is to build a **multivariable regression model** that predicts house prices using multiple features describing the area and property.

Instead of relying on a single variable, we now use **13 different predictors** to estimate price.

---

### What Features Influence House Prices?

The model will consider characteristics such as:

* Average number of rooms (`RM`)
* Distance to employment centres (`DIS`)
* Nitric oxide pollution levels (`NOX`)
* Percentage of lower-income population (`LSTAT`)
* Student–teacher ratio (`PTRATIO`)
* Proximity to the Charles River (`CHAS`)
* And several additional economic and environmental factors

---

### What You Will Accomplish Today

By the end of this lesson, you will:

1. Analyse and explore the Boston housing dataset
2. Split the data into training and testing sets
3. Train a multivariable linear regression model
4. Evaluate the model’s coefficients and residuals
5. Apply a log transformation to improve model performance
6. Use the trained model to estimate a property's market value

---

### Project Objective

The final goal is to turn your regression model into a **valuation engine**:

Property Features → Regression Model → Estimated Market Price

You will learn not just how to fit a model, but how to:

* Interpret coefficients economically
* Diagnose model quality using residual plots
* Improve model performance using transformations
* Make real-world price predictions
