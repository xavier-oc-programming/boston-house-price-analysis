### Key Takeaways

**1. Residual analysis is not optional**

Even when r² looks good, residual plots reveal whether the model's assumptions are met. A cone-shaped residual pattern signals heteroscedasticity — the model's errors are not uniform across the range of predicted values. This is a violation of OLS assumptions and suggests a transformation may help.

**2. Log-transforming the target is a standard fix**

When the target variable is right-skewed (as PRICE was), applying `np.log()` often:
- Reduces skew in the distribution
- Improves residual normality
- Lifts r² on both training and test sets

Remember to `np.exp()` the prediction to get back to dollar values.

**3. Coefficient signs tell the story**

| Feature | Coefficient sign | Interpretation |
|---------|-----------------|----------------|
| RM      | Positive        | More rooms → higher price |
| LSTAT   | Negative        | More poverty → lower price |
| CRIM    | Negative        | More crime → lower price |
| NOX     | Negative        | More pollution → lower price |
| CHAS    | Positive        | River access → price premium |
| DIS     | Positive        | Further from industry → higher price |

Always sense-check: if a sign is counterintuitive, investigate collinearity or data issues.

**4. Train/test split prevents overfit reporting**

Reporting only training r² is misleading — a model can memorise training data and perform poorly on new data. The test r² is the honest number to report.

### What to Remember

- Multivariable regression is powerful but assumes **linearity, independence, homoscedasticity, and normality of residuals**.
- When assumptions are violated, **transform** rather than ignore.
- The coefficients in a log-transformed model represent **percentage changes**, not absolute changes.
- sklearn's API is consistent: `.fit()` → `.score()` → `.predict()` works the same across all estimators.
