### Introduction

Welcome to Boston, Massachusetts in the 1970s.

Your task is to build a **multivariable regression model** that predicts a home's price based on neighbourhood characteristics such as:

### Why Multivariable Regression?

In simple linear regression you have one predictor:

$$\hat{PRICE} = \theta_0 + \theta_1 \cdot RM$$

In multivariable regression you use all available features simultaneously:

$$\hat{PRICE} = \theta_0 + \theta_1 RM + \theta_2 NOX + \theta_3 DIS + \theta_4 CHAS + \ldots + \theta_{13} LSTAT$$

Each coefficient $\theta_i$ represents the change in price for a one-unit increase in that feature, **holding all other features constant**.

### Analysis Flow

```
boston.csv
    ↓
[Load & inspect]   pd.read_csv(), .info(), .describe(), .isna()
    ↓
[Visualise]        sns.displot(), sns.jointplot(), sns.pairplot(), px.bar()
    ↓
[Split]            train_test_split() — 80% train / 20% test
    ↓
[Model v1]         LinearRegression().fit() → residual analysis
    ↓
[Transform]        np.log(PRICE) → reduced skew in residuals
    ↓
[Model v2]         LinearRegression().fit() on log target → r² improves
    ↓
[Valuation]        predict() on custom property → np.exp() to recover price
```

### Key Methods

| Method | Purpose |
|--------|---------|
| `pd.read_csv(path, index_col=0)` | Load dataset with row-number index |
| `df.describe()` | Summary statistics for all columns |
| `sns.displot(df[col], kde=True)` | Distribution with KDE overlay |
| `sns.jointplot(x=..., y=...)` | Bivariate scatter + marginals |
| `sns.pairplot(df)` | All pairwise scatter plots |
| `train_test_split(X, y, test_size=0.2)` | Split into training and test sets |
| `regr.fit(X_train, y_train)` | Train the model |
| `regr.score(X_test, y_test)` | Compute r² on held-out data |
| `regr.coef_` | Retrieve fitted coefficients |
| `np.log(y)` / `np.exp(y_pred)` | Log-transform target and invert |
